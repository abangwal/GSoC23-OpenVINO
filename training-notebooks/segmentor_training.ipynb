# Training Segmentor
Once we get the cropped image of the meter from the detector, the next step involves semantic segmentation of the cropped region to read scale and pointer.
Again, there are many approaches, and we will use the DL approach.


There are a lot of semantic segmentation models, and UNET is one of the early birds. UNET is based on encoder-decoder format and returns an image of the same dimensions as the input, with each pixel denoting a specific class; in our case, these classes will be background, scale, and pointer.


We will use a Python library called SegmentationModels, which contains the architecture of many famous segmentation models, including UNET. It also allows us to use pre-trained backbones like EfficientNet, ResNet, VGG, etc.

```python
import segmentation_models as sm
import albumentations as A

import cv2
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import os
import pathlib

sm.set_framework('tf.keras')
```

```python
# downlaoding PaddlePaddle's free meter segmentation dataset
seg = 'https://bj.bcebos.com/paddlex/examples/meter_reader/datasets/meter_seg.tar.gz'

if ('meter_seg' not in os.listdir()):
    for i in [seg]:
        !wget $i
        dir_name = i.split('/')[-1]
        !tar -xf $dir_name
else:
    print('Directory already exist')
```

## Image augmentation
We will use PaddlePaddle's free meter segmentation dataset, which contains 374 instances. We will also perform data augmentation to generate more synthetic data, for which we will use the `albumentation` library and, with its help, make our data 3x larger.

```python
augmentor = A.Compose([A.HorizontalFlip(p=0.5), 
                       A.ShiftScaleRotate(p=0.5),  # rotates image
                       A.Blur(blur_limit=3),       # blur image
                       A.ColorJitter(),            # augment Brightness , contrast , saturation
                       A.RGBShift()])              # RGB shifts
```

```python
augmented_data = pathlib.Path('augmented_data')
augmented_image = pathlib.Path('augmented_data/augmented_images')
augmented_label = pathlib.Path('augmented_data/augmented_labels')

augmented_data.mkdir(exist_ok=True)
augmented_image.mkdir(exist_ok=True)
augmented_label.mkdir(exist_ok=True)
```

```python
for i in sorted(os.listdir('/content/meter_seg/images/train')):
    img_path = '/content/meter_seg/images/train/' + i
    ann_path = '/content/meter_seg/annotations/train/' + i[:-3] +'png'

    for t in range(3):
        img = cv2.imread(img_path)
        ann = cv2.imread(ann_path)

        augmented = augmentor(image=img,mask=ann)
        augmented_img = augmented['image']
        augmented_ann = augmented['mask']

        cv2.imwrite(str(augmented_image)+ '/' + i[:-4] + str(t) + '.jpg', augmented_img)
        cv2.imwrite(str(augmented_label)+ '/' + i[:-4] + str(t) + '.png', augmented_ann)
```

## Training
First we load the augmented data since dataset is not too large we can use NumPy-ndarray as input, then initialized the model with SegmentationModel library and used `efficientnetb1` as model's backbone. Then trained it with Tensorflow's keras API

```python
def load_image(path,mask=False):
    images = []
    for i in sorted(os.listdir(path)):
        img = cv2.imread(path+'/'+i,0 if mask else 1)
        if not mask : img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB) 
        img = cv2.resize(img,(256,256),cv2.INTER_AREA)
        images.append(img)
    return np.array(images)
```

```python
train_image = load_image('/content/augmented_data/augmented_images')[:700]
train_ann = load_image('/content/augmented_data/augmented_labels',mask=True)[:700]
train_ann = tf.keras.utils.to_categorical(train_ann)
```

```python
rand_idx = np.random.randint(0,len(train_image))
plt.figure(figsize=(5,7))
plt.subplot(121);plt.imshow(train_image[rand_idx]);plt.axis(False);plt.title('image')
plt.subplot(122);plt.imshow(train_ann[rand_idx]);plt.axis(False);plt.title('annotation[RGB]')
plt.show()
```

```python
model = sm.Unet('efficientnetb1',classes=3,activation='softmax')
```

```python
model.summary()
```

```python
opt = tf.keras.optimizers.Adam()
loss = tf.keras.losses.CategoricalCrossentropy()
callback = tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=3)

model.compile(opt,loss)

his = model.fit(train_image,train_ann,batch_size=4,epochs=50,validation_split=0.2,callbacks=[callback])
```

```python
model.save('saved_model')
```

### Verifing inference
Once we have trained and saved the model, we can verify the inference by loading the model back into the TF-Keras Model class.

```python
val_image = load_image('/content/meter_seg/images/val')
val_ann = load_image('/content/meter_seg/annotations/val',mask=True)
```

```python
val_pred = model.predict(val_image,verbose=0.1)
val_pred = np.argmax(val_pred,axis=3)
```
